```python
from pathlib import Path
import subprocess

ROOT = Path(r'F:/workspace/ai/PhoenixAgent')
PY = ROOT / '.venv' / 'Scripts' / 'python.exe'

def run_cmd(args):
    result = subprocess.run(args, cwd=ROOT, capture_output=True, text=True, encoding='utf-8')
    print(result.stdout)
    if result.stderr:
        print(result.stderr)
    return result.returncode
```
